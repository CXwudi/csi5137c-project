## Preprocessing

```python
from modules.preprocessing import proprocessing_text
tokenized_dataset = proprocessing_text(preprocessed_dataset)
```

```python
from functools import reduce
tokenized_dataset["merged_comments"] = tokenized_dataset["comments"].apply(lambda x: reduce(lambda a, b: a + b, x, []))
tokenized_dataset
```

## To trainable dataset

1. select useful columns
2. apply Bag of Word with TF-IDF (instead of word count) on "title", "selftext", and "comments"
3. create 3 separate BoW, each for "title", "selftext", and "comments"

```python
trainable_dataset = tokenized_dataset[["no.Comments", "ups"]] # score column is same as ups, upratio is kind of wired because down is always 0, hence only no.Comments and ups are used
trainable_dataset
```

```python
from modules.vectorize import tf_idf_pd_serise, serise_to_BoW_dataframe
for text_column_name in ["title", "selftext", "merged_comments"]:
  scores = tf_idf_pd_serise(tokenized_dataset[text_column_name])
  bow = serise_to_BoW_dataframe(tokenized_dataset[text_column_name], scores, text_column_name + "_")
  trainable_dataset = pd.concat([trainable_dataset, bow], axis=1)
trainable_dataset
```

```python
trainable_dataset = pd.concat([trainable_dataset, tokenized_dataset["tag"]], axis=1)
trainable_dataset
```

```python
trainable_dataset.to_csv("../data/trainable_dataset.csv", index=False)
```

## Read and preprocess test dataset

```python
import ast
test_dataset = pd.read_csv("../data/our_test_data.csv")
test_dataset.fillna("", inplace=True)
test_dataset["comments"] = test_dataset["comments"].apply(lambda x: ast.literal_eval(x))
test_dataset["tag"] = test_dataset["tag"].apply(lambda x: int(x))
test_dataset
```

```python
tokenzied_test_dataset = proprocessing_text(test_dataset)
tokenzied_test_dataset["merged_comments"] = tokenzied_test_dataset["comments"].apply(lambda x: reduce(lambda a, b: a + b, x, []))
tokenzied_test_dataset
```

## To testable dataset

```python
testable_dataset = tokenzied_test_dataset[["no.Comments", "ups"]]
testable_dataset
```

```python
all_BoW_columns = pd.DataFrame()
for text_column_name in ["title", "selftext", "merged_comments"]:
  scores = tf_idf_pd_serise(tokenzied_test_dataset[text_column_name])
  bow = serise_to_BoW_dataframe(tokenzied_test_dataset[text_column_name], scores, text_column_name + "_")
  all_BoW_columns = pd.concat([all_BoW_columns, bow], axis=1)
all_BoW_columns
```

### Now we need to the select BoW column that appears in train dataset

And for column that appears in train dataset but not in test dataset, we need to add a column with all 0s.

```python
columns_in_trainable_dataset = list(trainable_dataset.columns[2:-1])
len(columns_in_trainable_dataset)
```

```python
columns_in_test_dataset = set(all_BoW_columns.columns)
filtered_BoW_columns = pd.DataFrame()
for col_name in columns_in_trainable_dataset:
  if col_name in columns_in_test_dataset:
    filtered_BoW_columns = pd.concat([filtered_BoW_columns, all_BoW_columns[col_name]], axis=1)
  else: 
    filtered_BoW_columns = pd.concat([filtered_BoW_columns, pd.DataFrame(columns=[col_name]).fillna(0)], axis=1)
# filtered_BoW_columns = all_BoW_columns[[col_name for col_name in columns_in_trainable_dataset if col_name in all_BoW_columns_name_set]]
# all_non_exist_columns = pd.DataFrame(columns=[col_name for col_name in columns_in_trainable_dataset if col_name not in all_BoW_columns_name_set])
# filtered_BoW_columns = pd.concat([filtered_BoW_columns, all_non_exist_columns], axis=1).fillna(0)
# filtered_BoW_columns
filtered_BoW_columns.fillna(0, inplace=True)
filtered_BoW_columns
```

```python
testable_dataset = pd.concat([testable_dataset, filtered_BoW_columns], axis=1)
testable_dataset = pd.concat([testable_dataset, tokenzied_test_dataset["tag"]], axis=1)
testable_dataset
```

```python
testable_dataset.to_csv("../data/testable_dataset.csv", index=False)
```

## Training part

```python
from sklearn.naive_bayes import GaussianNB,ComplementNB,BernoulliNB,CategoricalNB
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import svm
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.utils import compute_class_weight
from imblearn.over_sampling import SMOTE
```

```python
trainable_dataset['tag'].unique()
```

```python
X = trainable_dataset.iloc[:, 0:-1]
y = trainable_dataset.iloc[:, -1]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
class_weights = compute_class_weight(class_weight='balanced',classes=np.unique(y_train),y=y_train)
class_weights
```

In the previous block you can see that our dataset is not balanced. At First we will train the models with imbalanced dataset

### NB 

```python
# NB=GaussianNB()
NB=ComplementNB()
# NB=BernoulliNB()
# NB=CategoricalNB()
```

```python
y_pred = NB.fit(X_train, y_train).predict(X_test)
print("Number of mislabeled points out of a total %d points : %d"   % (X_test.shape[0], (y_test != y_pred).sum()))
score=NB.score(X_test,y_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

### RF

```python
RF=RandomForestClassifier(n_estimators=100)
RF.fit(X_train,y_train)
y_pred=RF.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

### SVM

```python

SVM = svm.SVC(kernel='linear', C=1, decision_function_shape='ovo') # Linear Kernel
SVM.fit(X_train, y_train)
y_pred = SVM.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

Training the models with balanced dataset 

### SMOTE

```python
sm = SMOTE(random_state = 2)
```

```python
X_train_res, y_train_res = sm.fit_resample(X_train, y_train.ravel())
```

```python
class_weights = compute_class_weight(class_weight='balanced',classes=np.unique(y_train_res),y=y_train_res)
class_weights
```

From here you can see that our dataset is now balanced dataset

### NB 

```python
# NB=GaussianNB()
NB_balance=ComplementNB()
# NB=BernoulliNB()
# NB=CategoricalNB()
```

```python
y_pred = NB_balance.fit(X_train_res, y_train_res).predict(X_test)
print("Number of mislabeled points out of a total %d points : %d"   % (X_test.shape[0], (y_test != y_pred).sum()))
score=NB_balance.score(X_test,y_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

### RF

```python
RF_balance=RandomForestClassifier(n_estimators=100)
RF_balance.fit(X_train_res,y_train_res)
y_pred=RF_balance.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

### SVM

```python

SVM_balance = svm.SVC(kernel='linear', C=1, decision_function_shape='ovo') # Linear Kernel
SVM_balance.fit(X_train_res, y_train_res)
y_pred = SVM_balance.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
