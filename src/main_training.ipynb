```python
from functools import reduce
tokenized_dataset["merged_comments"] = tokenized_dataset["comments"].apply(lambda x: reduce(lambda a, b: a + b, x, []))
tokenized_dataset
```

## To trainable dataset

1. select useful columns
2. apply Bag of Word with TF-IDF (instead of word count) on "title", "selftext", and "comments"
3. create 3 separate BoW, each for "title", "selftext", and "comments"

```python
trainable_dataset = tokenized_dataset[["no.Comments", "ups"]] # score column is same as ups, upratio is kind of wired because down is always 0, hence only no.Comments and ups are used
trainable_dataset
```

```python
from modules.vectorize import tf_idf_pd_serise, serise_to_BoW_dataframe
for text_column_name in ["title", "selftext", "merged_comments"]:
  scores = tf_idf_pd_serise(tokenized_dataset[text_column_name])
  bow = serise_to_BoW_dataframe(tokenized_dataset[text_column_name], scores, text_column_name + "_")
  trainable_dataset = pd.concat([trainable_dataset, bow], axis=1)
trainable_dataset
```

```python
trainable_dataset = pd.concat([trainable_dataset, tokenized_dataset["tag"]], axis=1)
trainable_dataset
```

## Read and preprocess test dataset

```python
import ast
test_dataset = pd.read_csv("../data/our_test_data.csv")
test_dataset.fillna("", inplace=True)
test_dataset["comments"] = test_dataset["comments"].apply(lambda x: ast.literal_eval(x))
test_dataset["tag"] = test_dataset["tag"].apply(lambda x: int(x))
test_dataset
```

```python
tokenzied_test_dataset = proprocessing_text(test_dataset)
tokenzied_test_dataset["merged_comments"] = tokenzied_test_dataset["comments"].apply(lambda x: reduce(lambda a, b: a + b, x, []))
tokenzied_test_dataset
```

## To testable dataset

```python
testable_dataset = tokenzied_test_dataset[["no.Comments", "ups"]]
testable_dataset
```

```python
all_BoW_columns = pd.DataFrame()
for text_column_name in ["title", "selftext", "merged_comments"]:
  scores = tf_idf_pd_serise(tokenzied_test_dataset[text_column_name])
  bow = serise_to_BoW_dataframe(tokenzied_test_dataset[text_column_name], scores, text_column_name + "_")
  all_BoW_columns = pd.concat([all_BoW_columns, bow], axis=1)
all_BoW_columns
```

### Now we need to the select BoW column that appears in train dataset

And for column that appears in train dataset but not in test dataset, we need to add a column with all 0s.

```python
columns_in_trainable_dataset = list(trainable_dataset.columns[2:-1])
len(columns_in_trainable_dataset)
```

```python
all_BoW_columns_name_set = set(all_BoW_columns.columns)
filtered_BoW_columns = all_BoW_columns[[col_name for col_name in columns_in_trainable_dataset if col_name in all_BoW_columns_name_set]]
all_non_exist_columns = pd.DataFrame(columns=[col_name for col_name in columns_in_trainable_dataset if col_name not in all_BoW_columns_name_set])
filtered_BoW_columns = pd.concat([filtered_BoW_columns, all_non_exist_columns], axis=1).fillna(0)
filtered_BoW_columns
```
